```python
loss_areas = pd.read_csv('loss_areas.csv')
```

```python
# loss_areas['sq_mi'] = loss_areas['sq_mi'].str.replace(',' ,'')
# loss_areas["sq_mi"] = loss_areas.sq_mi.astype(float)
```

```python
# mile_to_ha = 258.99
```

```python
loss_areas['area_ha'] = loss_areas['sq_mi'] * 258.99
```

```python
loss_areas
```

```python
loss_areas['pct_area'] = (loss_areas['umd_tree_cover_loss__ha']/loss_areas['area_ha'])*100
```

```python
loss_areas
```

```python
loss_areas.sort_values(by='pct_area', ascending = False).head(10)
```

```python
(
    ggplot(loss_areas.sort_values(by='pct_area', ascending = False).head(10))
    + aes(x='name', y='umd_tree_cover_loss__ha')
    + geom_bar(stat='identity')
#     + coord_flip()
    + theme(figure_size=(16, 12))
)
```

```python
model = smf.ols("gfw_gross_emissions_co2e_all_gases__Mg ~ umd_tree_cover_loss__ha", data=loss_areas)
results = model.fit()
results.summary()
```
