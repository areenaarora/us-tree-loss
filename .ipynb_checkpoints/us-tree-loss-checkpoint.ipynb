```python
import pandas as pd
df = pd.read_csv('treecover_loss__ha.csv')
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("-", "_")
df.columns = df.columns.str.lower()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df.shape
```

```python
from plotnine import *
```

```python
df
```

Timeline of treecover loss in the US

```python
(
    ggplot(df)
    + aes(x='umd_tree_cover_loss__year', y='umd_tree_cover_loss__ha')
    + geom_bar(stat='identity')
#     + coord_flip()
)
```

```python
(
    ggplot(df)
    + aes(x='umd_tree_cover_loss__year', y='umd_tree_cover_loss__ha')
    + geom_line()
)
```

Adding another dataset for regional tree cover loss data; from the same source.

```python
regions = pd.read_csv('us_loss_regions.csv')
```

```python
regions
```

```python
regions.groupby(by='name').reset_index()
```

### Top five states with the highest emission levels in 2020

```python
(
    ggplot(regions.loc[regions['umd_tree_cover_loss__year'] == 2020].sort_values(by='gfw_gross_emissions_co2e_all_gases__Mg', ascending= False).head(5))
    + aes(x='name', y='gfw_gross_emissions_co2e_all_gases__Mg')
    + geom_bar(stat='identity')
    + coord_flip()
    + theme(figure_size=(16, 12))
)
```

```python
regions.loc[regions['umd_tree_cover_loss__year'] == 2020]
```

#### Same chart but with top 10 states

```python
(
    ggplot(regions.loc[regions['umd_tree_cover_loss__year'] == 2020].sort_values(by='gfw_gross_emissions_co2e_all_gases__Mg', ascending= False).head(10))
    + aes(x='name', y='gfw_gross_emissions_co2e_all_gases__Mg')
    + geom_bar(stat='identity')
#     + coord_flip()
    + theme(figure_size=(16, 12))
)
```

```python
regions.loc[regions['umd_tree_cover_loss__year'] == 2020].sort_values(by='gfw_gross_emissions_co2e_all_gases__Mg', ascending= False).head(10)
```

#### Top 10 states by tree cover loss

```python
(
    ggplot(regions.loc[regions['umd_tree_cover_loss__year'] == 2001].sort_values(by='umd_tree_cover_loss__ha', ascending = False).head(10))
    + aes(x='name', y='umd_tree_cover_loss__ha')
    + geom_bar(stat='identity')
#     + coord_flip()
    + theme(figure_size=(16, 12))
)
```
